### Zorder
In Matplotlib, the zorder parameter is used to control the drawing order of different plot elements. The zorder value determines which elements are drawn on top of others. Elements with higher zorder values will be drawn above those with lower values.


````python
fig = plt.figure(figsize=(7.5, 7.5))
ax = fig.add_axes([0.2, 0.17, 0.68, 0.7], aspect=1)
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)



X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

ax.plot(X, Y1, c='C0', lw=2.5, label="Blue signal", zorder=1)
ax.plot(X, Y1, c='y', lw=8, label="Yellow signal", zorder=0) # Zorder with low value goes to bottom.
ax.plot(X, Y2, c='C1', lw=2.5, label="Orange signal")
ax.plot(X[::3], Y3[::3], linewidth=0, markersize=9,
        marker='s', markerfacecolor='none', markeredgecolor='C4',
        markeredgewidth=1.5)

plt.show()
````

## 5. Annotation

Let's create a function to add custom annotations to our plot.

````python
fig = plt.figure(figsize=(7.5, 7.5))
ax = fig.add_axes([0.2, 0.17, 0.68, 0.7], aspect=1)
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)


royal_blue = [0, 20/256, 82/256]

def annotate(x, y, text, code):
    # Circle marker
    c = Circle((x, y), radius=0.15, clip_on=False, zorder=10, linewidth=2.5,
               edgecolor=royal_blue + [0.6], facecolor='none',
               path_effects=[withStroke(linewidth=7, foreground='white')])
    ax.add_artist(c)

    # Text and code with path effects
    for path_effects in [[withStroke(linewidth=7, foreground='white')], []]:
        color = 'white' if path_effects else royal_blue
        ax.text(x, y-0.2, text, zorder=100,
                ha='center', va='top', weight='bold', color=color,
                style='italic', fontfamily='monospace',
                path_effects=path_effects)

        color = 'white' if path_effects else 'black'
        ax.text(x, y-0.33, code, zorder=100,
                ha='center', va='top', weight='normal', color=color,
                fontfamily='monospace', fontsize='medium',
                path_effects=path_effects)

# Add annotations
annotate(3.5, -0.13, "Minor tick label", "ax.xaxis.set_minor_formatter")
annotate(-0.03, 1.0, "Major tick", "ax.yaxis.set_major_locator")
annotate(0.00, 3.75, "Minor tick", "ax.yaxis.set_minor_locator")
annotate(-0.15, 3.00, "Major tick label", "ax.yaxis.set_major_formatter")
annotate(1.68, -0.39, "xlabel", "ax.set_xlabel")
annotate(-0.38, 1.67, "ylabel", "ax.set_ylabel")
annotate(1.52, 4.15, "Title", "ax.set_title")
annotate(1.75, 2.80, "Line", "ax.plot")
annotate(2.25, 1.54, "Markers", "ax.scatter")
annotate(3.00, 3.00, "Grid", "ax.grid")
annotate(3.60, 3.58, "Legend", "ax.legend")
annotate(2.5, 0.55, "Axes", "fig.subplots")
annotate(4, 4.5, "Figure", "plt.figure")
annotate(0.65, 0.01, "x Axis", "ax.xaxis")
annotate(0, 0.36, "y Axis", "ax.yaxis")
annotate(4.0, 0.7, "Spine", "ax.spines")

plt.show()
````

### Circular plot

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle
from matplotlib.patheffects import withStroke

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Define circle properties
x, y = 0, 0.5  # Center of the circle
royal_blue = (65/255, 105/255, 225/255)  # Define royal blue color

# Create a Circle
c = Circle((x, y), radius=0.1, clip_on=False, zorder=10, linewidth=2.5,
            edgecolor=royal_blue + (0.6,), facecolor='pink',
            path_effects=[withStroke(linewidth=10, foreground='y')])

# Add the circle to the axis
ax.add_artist(c)

# Set limits and aspect
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal', adjustable='box')

# Optional: Add grid and title
ax.grid(True)
ax.set_title("Circle with Custom Edge Properties")

# Show the plot
plt.show()
````

````python
#Start your solutions from this cell

# Create data
x = np.linspace(0, 10, 100)
y1 = 4 + 2 * np.sin(2 * x)
y2 = 4 + 2 * np.cos(2 * x)

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot data
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# Modify the x-axis tick labels to show π multiples instead of numbers.
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π', '5π/2'])
ax.set_xticks([1, 3, 5, 7, 9], minor = True)

#Add a grid to the plot with dashed lines. Make the major grid lines darker and thicker than the minor grid lines.
ax.grid(True, which='both')
ax.grid(which='major', color = 'black', linestyle='--', linewidth=1)
ax.grid(which='minor', color = 'gray', linestyle='--', linewidth=0.5)

#3 customise spines
# Move the left spine to the center (x = 0)
#ax.spines['left'].set_position('center')

# Move the bottom spine to the center (y = 0)
#ax.spines['bottom'].set_position('center')

# Hide the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#4 annotate  -- Add a text annotation pointing out the maximum point of the sine wave.
ax.annotate('sin- max', xy=(x[np.argmax(y1)], y1[np.argmax(y1)]), xytext=(x[np.argmax(y1)-1], y1[np.argmax(y1)]-1), 
            color = 'red',
            arrowprops=dict(facecolor="blue", arrowstyle='->'))


# 5 secondary y-axis Add a secondary y-axis on the right side of the plot, showing the values in radians (original values divided by pi
def to_radians(y):
    return y / np.pi

def from_radians(y):
    return y * np.pi

secy = ax.secondary_yaxis('right', functions = (to_radians, from_radians)) # functions = (converter , inverse function of converter)
secy.set_ylabel('radians')
# Set title and labels
ax.set_title("Trigonometric Functions")
ax.set_xlabel("x")
ax.set_ylabel("y")

inset_ax = ax.inset_axes([0.2, 1, 0.2, 0.75])
# Plot a zoomed-in section of the sine wave in the inset
x_inset = np.linspace(3.75, 4.25, 100)
y_inset = 4 + 2 * np.sin( 2 * x_inset )
inset_ax.plot(x_inset, y_inset, color='green')
ax.indicate_inset_zoom(inset_ax, edgecolor='purple', linewidth=2)

# Add legend
ax.legend()

# Change the overall style of the figure
plt.style.use('ggplot')  # You can use any style like 'ggplot', 'seaborn', etc.

# Define pastel colors
pastel_colors = ['#FFB3BA', '#FFDFBA', '#FFFFBA', '#BAFFC9', '#BAE1FF']

# Modify the color cycle to use pastel colors
plt.rcParams['axes.prop_cycle'] #= cycler(color=pastel_colors)

plt.show()
````
