# Car Sales Datase

The `Car_Sales.csv` dataset contains 1000 rows of car sales data. Here's is the breakdown of the attributes of the dataset:

`date`: covering about 2.7 years.

`make`: car manufacturers.

`model`: models for each manufacturer.

`sales`: daily sales numbers.

`price`: in pounds.

`mpg`:  miles per gallon.

`type`: type of car,  `Sedan`, `SUV`, `Truck`, or `Coupe`.

```python
#First, let's import the necessary libraries:
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For creating static, animated, and interactive visualizations
import seaborn as sns  # For statistical data visualization
```

## 1. Reading A Dataset

In this section, we'll read a dataset to `df` (dataframe).

```python
# Load the dataset from the CSV file in the same folder
df = pd.read_csv('Car_Sales.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Display basic information about the dataset (columns, data types, non-null count)
print(df.info())
```

```python
# Display summary statistics of the numerical columns
print(df.describe())
```

```python
# Check for missing values in each column
print(df.isnull().sum())
```

## 2. Data Preparation


```python
# Convert 'date' column to datetime type for time-based analysis
df['date'] = pd.to_datetime(df['date'])
```

## 3.  Basic Visualisations


```python
# Create a bar plot of average sales by make
plt.figure(figsize=(12, 6))  # Set the figure size
df.groupby('make')['sales'].mean().sort_values(ascending=False).plot(kind='bar')  # Group by make, calculate mean sales, sort, and plot
plt.title('Average Sales by Make')  # Set the title
plt.xlabel('Make')  # Set the x-axis label
plt.ylabel('Average Sales')  # Set the y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust the layout
plt.savefig('average_sales_by_make.png')  # Save the plot as a PNG file
plt.savefig('average_sales_by_make.pdf')  # Save the plot as a PDF file
plt.show()  # Display the plot
```

```python
# Create a scatter plot of price vs. mpg
plt.figure(figsize=(10, 6))  # Set the figure size
sns.scatterplot(data=df, x='price', y='mpg', hue='type')  # Create a scatter plot with price on x-axis, mpg on y-axis, colored by type
plt.title('Price vs. MPG by Vehicle Type')  # Set the title
plt.xlabel('Price')  # Set the x-axis label
plt.ylabel('Miles Per Gallon (MPG)')  # Set the y-axis label
plt.tight_layout()  # Adjust the layout
plt.savefig('price_vs_mpg.png')  # Save the plot as a PNG file
plt.savefig('price_vs_mpg.pdf')  # Save the plot as a PDF file
plt.show()  # Display the plot
```

## 1. Total Sales for Each Car

```python
# Calculate and display the total sales for each car make.
sales = df.groupby('make')['sales'].sum().reset_index()
sales.columns = ['make', 'sales']
```

```python
sales
```

## 2. Car the Highest Price

Find the car model with the highest average price

```python
avg_price = df.groupby('model')['price'].mean().reset_index()
avg_price.columns = ['model', 'avg_price']
```

```python
avg_price.sort_values(by='avg_price', ascending=False).head(1)
```

## 3. Find Top Selling Cars

Find the top 5 best-selling car models.

```python
best_selling = df.groupby('model')['sales'].sum().reset_index()
best_selling.columns = ['model', 'Total Sales']
```

```python
best_selling.sort_values(by='Total Sales', ascending=False).head(5)
```

## 4. histogram of the Sales

Create a histogram of the sales column. Save the plot as sales_distribution.png and sales_distribution.pdf.

What does the distribution look like?

```python
plt.hist(df['sales'])
plt.xlabel('sales')
plt.ylabel('Count')
plt.title('Histogram of sales')
plt.tight_layout()  # Adjust the layout
plt.savefig('sales_distribution.png')  # Save the plot as a PNG file
plt.savefig('sales_distribution.pdf')  # Save the plot as a PDF file
plt.show()  # Display the plot
```

```python
df['sales'].value_counts()
```

## Average of Type

Calculate the average price for each vehicle type and visualise it using a bar plot.

Save the plot as avg_price_by_type.png and avg_price_by_type.pdf.

```python
df.groupby('type')['price'].mean().plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Price of vehicle Types')
plt.xlabel('Type')
plt.ylabel('Price')
plt.savefig('avg_price_by_type.png')
plt.savefig('avg_price_by_type.pdf')
plt.show()
```

## 6. Average of Type

Create a line plot showing the trend of total sales over time.

Save the plot as sales_trend.png.

```python
plt.figure(figsize=(15, 6))  # Set the figure size
plt.plot(df['date'], df['sales'])
plt.xticks(rotation=90)
```

## 6. Challenging Task: Plot Matrix

Create a scatter plot matrix for the numerical variables (sales, price, mpg). Save the plot as scatter_matrix.png.

```python
pd.plotting.scatter_matrix(df[['sales', 'price', 'mpg']], figsize=(12, 10), diagonal='kde', marker='o', alpha=0.7, color='blue')
# Adding a title to the entire figure
plt.suptitle('Matrix plot of "sales, price, mpg"', size=16)
plt.savefig('scatter_matrix.png')
plt.show()
```

## 7. Challenging Task: Perform a Year-Over-Year Analysis of Sales

Perform a year-over-year analysis of sales.

Calculate the total sales for each year and visualise the results.

Save the plot as yearly_sales.png.

```python
df['year'] = df['date'].dt.year
yearly_sales = df.groupby('year')['sales'].sum()
```

```python
yearly_sales
```

```python
# Plot the total sales per year
plt.figure(figsize=(8, 6))
yearly_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Year-over-Year Total Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# Save the plot as PNG
plt.savefig('yearly_sales.png')

# Show the plot
plt.show()
```
